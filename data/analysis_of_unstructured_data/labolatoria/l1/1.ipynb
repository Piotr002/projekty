```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
data = pd.read_csv("Titanic-Dataset.csv")
data.head()
```

```python
data.columns
```

```python
plt.figure(figsize=(10, 10))
plt.pie(data.Sex.value_counts(), labels=["male", "female"])
plt.legend()
```

```python
plt.figure(figsize=(16, 12))
for i in data.Sex.unique():
    temp = data.query(f'Sex=="{i}"')
    plt.scatter(temp.Age, temp.Fare)
plt.legend(data.Sex.unique())
```
