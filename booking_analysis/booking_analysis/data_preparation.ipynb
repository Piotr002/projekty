```python
import pandas as pd
import ast
```

```python
data = pd.read_csv("hotel_dataset.csv", index_col=0)
```

```python
data.guest_opinions[7083].split("\r\n")
```

```python
guest_opinions = [{} for i in range(len(data.guest_opinions))]
for i in range(1, len(data.guest_opinions) + 1):
    temp = data.guest_opinions[i].split("\r\n")
    for j in range(len(temp)):
        if temp[j] == "Kategorie:":
            start = j
        elif temp[j] == 'Wybierz tematy, aby przeczytać opinie:' or temp[
            j] == 'Zobacz, co najbardziej podobało się Gościom:':
            end = j
            temp = temp[start:end]
            break
    for j in range(len(temp)):
        if temp[j] == "Personel":
            guest_opinions[i - 1]["Personel"] = float(temp[j + 1].replace(",", "."))
        elif temp[j] == "Udogodnienia":
            guest_opinions[i - 1]["Udogodnienia"] = float(temp[j + 1].replace(",", "."))
        elif temp[j] == "Czystość":
            guest_opinions[i - 1]["Czystość"] = float(temp[j + 1].replace(",", "."))
        elif temp[j] == "Komfort":
            guest_opinions[i - 1]["Komfort"] = float(temp[j + 1].replace(",", "."))
        elif temp[j] == "Stosunek jakości do ceny":
            guest_opinions[i - 1]["Stosunek jakości do ceny"] = float(temp[j + 1].replace(",", "."))
        elif temp[j] == "Lokalizacja":
            guest_opinions[i - 1]["Lokalizacja"] = float(temp[j + 1].replace(",", "."))
        elif temp[j] == "Bezpłatne WiFi":
            guest_opinions[i - 1]["Bezpłatne WiFi"] = float(temp[j + 1].replace(",", "."))
```

```python
capital_letters = "AĄBCĆDEĘFGHIJKLŁMNŃOÓPRQUSŚTÓWXYZŹŻ"
hotel_amenities = [{} for i in range(len(data.guest_opinions))]
for i in range(1, len(data.amenities) + 1):
    temp = data.amenities[i].split("\r\n")
    for j in range(len(temp)):
        if "Najpopularniejsze udogodnienia" == temp[j]:
            index = 2
            start = j
            break
    while (temp[start + index][0] not in capital_letters and "śniadanie" not in temp[start + index].lower()) or "Wi-Fi" in temp[start+index]:
        index += 1
    hotel_amenities[i - 1]["Najpopularniejsze udogodnienia"] = temp[start + 1:start + index + 1] if "śniadanie" in temp[start + index].lower() else temp[start + 1:start + index]
for i in range(1, len(data.amenities) + 1):
    temp = data.amenities[i].split("\r\n")
    if "Łazienka" not in temp:
        continue
    for j in range(len(temp)):
        if "Łazienka" == temp[j]:
            index = 1
            start = j
            break
    while temp[start + index][0] not in capital_letters:
        index += 1
    hotel_amenities[i - 1]["Łazienka"] = temp[start + 1:start + index]

for i in range(1, len(data.amenities) + 1):
    temp = data.amenities[i].split("\r\n")
    if "Media/Technologia" not in temp:
        continue
    for j in range(len(temp)):
        if "Media/Technologia" == temp[j]:
            index = 1
            start = j
            break
    while temp[start + index][0] not in capital_letters:
        index += 1
    hotel_amenities[i - 1]["Media/Technologia"] = temp[start + 1:start + index]
    
for i in range(1, len(data.amenities) + 1):
    temp = data.amenities[i].split("\r\n")
    if "Języki" not in temp or temp.index("Języki") == len(temp) - 1:
        continue
    for j in range(len(temp)):
        if "Języki" == temp[j]:
            index = 1
            start = j
            break
    while temp[start + index][0] not in capital_letters:
        if start + index == len(temp) - 1:
            break
        index += 1
    hotel_amenities[i - 1]["Języki"] = temp[start + 1:start + index]
```
